```python
class ATM:
    def __init__(self):
        self.pin=1111
        self.balance=5000

    def change_pin(self):
        self.pin=input('enter new input')
        print('you PIN has been changed')

    def show_balance(self):
        print(f'''your current balance is rs{self.balance}''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
axis=ATM()
```

```python
axis.show_balance()
```
